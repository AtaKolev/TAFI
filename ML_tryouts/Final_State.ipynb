```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
```

```python
start = '2015-01-05'
end = '2020-09-04'

eur_df = yf.download('EURUSD=X', start, end)
```

```python
eur_df.drop('Volume', axis = 1, inplace = True)
```

```python
eur_df.head()
```

```python
open_mean = []
high_mean = []
low_mean = []
close_mean = []

for i in range(len(eur_df['Adj Close'])):
    open_mean.append(eur_df['Open'][i-5:i].mean())
    high_mean.append(eur_df['High'][i-5:i].mean())
    low_mean.append(eur_df['Low'][i-5:i].mean())
    close_mean.append(eur_df['Adj Close'][i-5:i].mean())
```

```python
data_dict = {'Open Mean' : open_mean,
             'High Mean' : high_mean,
             'Low Mean' : low_mean,
             'Adj Close Mean' : close_mean}
```

```python
df_mean = pd.DataFrame(data_dict)
```

```python
df_mean
```

```python
df_mean.dropna()
```

```python
def price_list(price):
    truth_list = []
    
    for i in range(len(price)):
        if i < len(price):
            if price[i-1] < price[i]:
                truth_list.append(1)
            else:
                truth_list.append(0)
    
    return truth_list
```

```python
list_of_truths = price_list(close_mean)
list_of_truths
```

```python
df_mean['Price Up'] = list_of_truths
```

```python
df_mean.dropna(inplace = True)
```

```python
df_mean.tail()
```

```python
df_mean['Price Up'] = df_mean['Price Up'].shift(periods = -1, fill_value = 1)
```

```python
X = df_mean.iloc[:, 0:-1].values
y = df_mean.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

lr = LogisticRegression()

lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print("CM: ", confusion_matrix(y_test, y_pred))
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```
