```python
def price_check(price):
    #Price must be Nx5 Matrix
    
    price_go_up_list = []
    
    for i in range(len(price)):
        if price[i] > price[i - 1]:
            price_go_up_list.append(1)
        else:
            price_go_up_list.append(0)
            
    return price_go_up_list
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
```

```python
start = '2015-05-01'
end = '2020-09-11'

tsla = yf.download('TSLA', start, end)
```

```python
mean_open = []
mean_high = []
mean_low = []
mean_AdjClose = []

for i in range(len(tsla['Open'])):
    mean_open.append(tsla['Open'][i])
    mean_high.append(tsla['High'][i])
    mean_low.append(tsla['Low'][i])
    mean_AdjClose.append(tsla['Adj Close'][i])
```

```python
tsla_means = pd.DataFrame()
```

```python
mean_op = []
mean_hi = []
mean_lo = []
mean_clo = []

for i in range(5, len(mean_open)):
    mean_op.append(tsla['Open'][i-5:i].mean())
    mean_hi.append(tsla['High'][i-5:i].mean())
    mean_lo.append(tsla['Low'][i-5:i].mean())
    mean_clo.append(tsla['Adj Close'][i-5:i].mean())
    

data_mean = {'Open mean' : mean_op,
             'High mean' : mean_hi,
             'Low mean' : mean_lo,
             'Adj Close Mean' : mean_clo}
```

```python
tsla_means = pd.DataFrame(data_mean)
```

```python
tsla_means.head()
```

```python
truth_list = price_check(mean_clo)
truth_list
```

```python
tsla_means['Price go UP'] = truth_list
```

```python
tsla_means
```

```python
X = tsla_means.iloc[:, 0:-1].values
y = tsla_means.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = lr.predict(X_test)

print('Confusion Matrix:', confusion_matrix(y_test, y_pred))
print("Accuracy Score:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred_dtc = dtc.predict(X_test)

print('Confusion Matrix DTC:', confusion_matrix(y_test, y_pred_dtc))
print('Accuracy Score DTC:', accuracy_score(y_test, y_pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 1000)

rfc.fit(X_train, y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)

print("Confusion Matrix RFC: ", confusion_matrix(y_test, y_pred_rfc))
print('Accuracy Score RFC: ', accuracy_score(y_test, y_pred_rfc))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(X_train, y_train)
y_pred_gnb = gnb.predict(X_test)

print('Confusion Matrix: ', confusion_matrix(y_test, y_pred_gnb))
print('Accuracy Score: ', accuracy_score(y_test, y_pred_gnb))
```
