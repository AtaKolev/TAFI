```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib inline
```

```python
start = '2015-01-05'
end = '2020-09-11'

tsla = yf.download('TSLA', start, end)
```

```python
def price_check(price):
    #Price must be Nx5 Matrix
    
    price_go_up_list = []
    
    for i in range(len(price)):
        if sum(price[i]) > sum(price[i - 1]):
            price_go_up_list.append(1)
        else:
            price_go_up_list.append(0)
            
    return price_go_up_list
```

```python
close_conv = []

for i in range(len(tsla['Adj Close'])):
    
    close_conv.append(tsla['Adj Close'][i])
```

```python
X = []

for i in range(5, len(close_conv)):
    X.append(close_conv[i-5:i])
```

```python
list_of_ups_tsla = price_check(X)
```

```python
list_of_ups_tsla
```

```python
y = list_of_ups_tsla
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

lr = LogisticRegression()

lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

print('Confusion Matrix: ', confusion_matrix(y_test, y_pred))
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```
